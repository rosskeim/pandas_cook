```python
import datetime
```

```python
import pandas_datareader.data as web
```

```python
import requests_cache
```

```python
session = requests_cache.CachedSession(
    cache_name="cache",
    backend="sqlite",
    expire_after=datetime.timedelta(days=90)
)
```

```python
tsla = web.DataReader(
    "tsla",
    data_source="yahoo",
    start="2017-1-1",
    session=session
)
```

```python
tsla.head(8)
```

```python
tsla_close = tsla["Close"]
```

```python
tsla_cummax = tsla_close.cummax()
```
